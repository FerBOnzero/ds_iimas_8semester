# 1. El algoritmo del descenso del gradiente

El descenso del gradiente es un algoritmo de optimización fundamental utilizado para encontrar el mínimo de una función. Es especialmente útil en problemas donde el cálculo del mínimo analítico es complejo o imposible. El algoritmo se mueve iterativamente hacia el mínimo de la función, ajustando los parámetros en dirección opuesta al gradiente de la función en el punto actual.

# 2. Función a minimizar

Vamos a trabajar con la función $f(x, y) = x² + y²$. El mínimo global de esta función se encuentra en el origen $(0, 0)$; matemáticamente, se puede encontrar este mínimo obteniendo el gradiente de la función e igualándolo a $0$. El gradiente de $f(x,y)$ es un vector compuesto por las derivadas parciales de $f$ con respecto a $x$ y $y$. En este caso, las derivadas parciales son:

$$\frac{\partial f}{\partial x} = 2x$$
$$\frac{\partial f}{\partial y} = 2y$$

Igualando a cero estas ecuaciones:

$$2x = 0$$
$$2y = 0$$

Resolviendo estas ecuaciones, obtenemos $x=0$ y $y=0$. Por lo tanto, el mínimo de la función $f(x, y)$ se encuentra en el punto $(0,0)$.

# 3. Implementación del algoritmo del descenso del gradiente


Vamos a implementar el algoritmo de descenso del gradiente para esta función. Empezaremos con un punto aleatorio y nos moveremos hacia el míni de forma iterativamo.

```python
from typing import Tuple, List

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)  # Resultados consistentes
```

```python
# Función f(x, y) = x² + y²
def f(x: float, y: float) -> float:
    return x**2 + y**2 


# Gradiente de f(x, y)
def grad_f(x: float, y: float) -> Tuple[float, float]:
    df_dx = 2 * x
    df_dy = 2 * y
    return df_dx, df_dy

def gradient_descent(x: float, y: float, learning_rate: float=0.1, iterations: int=100) -> Tuple[float, float]:
    for _ in range(iterations):
        df_dx, df_dy = grad_f(x, y)
        x -= learning_rate * df_dx
        y -= learning_rate * df_dy
    return x, y

# Parámetros iniciales
x_initial, y_initial = np.random.uniform(-10, 10, 2)

# Llamada a la función
min_x, min_y = gradient_descent(x_initial, y_initial)
print(f"El mínimo se encuentra en: ({min_x}, {min_y})")
```

# 4. Implementación de variantes del algoritmo del decenso del gradiente

Recuerda que existen las siguientes variantes del algoritmo del decenso del gradiente:

- Descenso del gradiente por lotes (batch gradient descent): Este método utiliza todo el conjunto de datos para calcular el gradiente en cada paso. Es computacionalmente intensivo, especialmente con grandes conjuntos de datos, pero los pasos son más precisos hacia el mínimo.
- Descenso del gradiente estocástico (stochastic gradient descent):  A diferencia del método por lotes, SGD actualiza los parámetros utilizando solo un punto de datos en cada iteración. Esto hace que cada paso sea menos preciso y más ruidoso, pero a menudo converge más rápidamente y puede manejar conjuntos de datos grandes de manera más eficiente.
- Descenso del gradiente por mini-lotes (mini-batch gradient descent): Este método es un equilibrio entre los dos anteriores. Utiliza un subconjunto pequeño pero significativo del conjunto de datos para calcular cada actualización. Ofrece una buena compensación entre la eficiencia computacional y la precisión de los pasos.

Tu tarea es ajustar un modelo de función seno en la forma $f(x) = sin(ax)$, esto es, el objetivo es estimar la frecuencia más precisa. Para medir el error, utiliza el error cuadrático medio MSE. Recuerda que:

$$MSE = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_1)^2$$

## 4.2 Definición del Modelo y la Función de Pérdida

```python
def model(x: np.ndarray, a: float) -> np.ndarray:
    """Genera las predicciones del modelo.
    :param x: Arreglo de NumPy de dimensiones (100,). Representa los puntos a evaluar.
    :param a: Frecuencia de la función seno.
    :return: Arreglo de NumPy de dimensiones (100,) con la evaluación de cada uno de los puntos de entrada con la frecuencia dada.
    """
    # TODO: Completa la función.
    return np.sin(a*x)


def mse(y_data: np.ndarray, y_pred: np.ndarray) -> float:
    """Calcula el error cuadrático medio.
    :param y_pred: Arreglo de NumPy de dimensiones (100, 0). Representa las predicciones hechas por el modelo.
    :param y_data: Arreglo de NumPy de dimensiones (100, 0). Representa los valores observados.
    :return: El error cuadrático medio.
    """
    # TODO: Completa la función
    sum_error = 0.0
    len_y = y_data.shape[0]
    for i in range(len_y):
        sum_error += (y_data[i] - y_pred[i])**2
        
    return 1/len_y * (sum_error)
```

## 4.3 Implementación del algoritmo del descenso del gradiente por lotes

```python
# Con esta función calcularemos el gradiente de la función de costo MSE con respecto al parámetro a.
def grad_mse(x, y_data, a):
    h = 1e-4  # un pequeño valor para la aproximación numérica
    mse_plus_h = mse(model(x, a + h), y_data)
    mse_minus_h = mse(model(x, a - h), y_data)
    grad = (mse_plus_h - mse_minus_h) / (2 * h)
    return grad
```

```python
def optimize(
    x: np.ndarray,
    y_data: np.ndarray,
    a: float,
    learning_rate: float=0.01,
    iterations=20
) -> Tuple[float, List[float]]:
    """Optimiza la función utilizando el algoritmo del descenso del gradiente por lotes.
    :param x: Arreglo de NumPy de dimensiones (100, 0). Representa los puntos a evaluar.
    :param y_data: Arreglo de NumPy de dimensiones (100, 0). Representa los valores observados.
    :param a: Frecuencia de la función seno.
    :param learning_rate: Tasa de aprendizaje.
    :param iterations: Número de iteraciones por optimizar.
    :return: Una tupla con dos valores, el primero es el valor óptimo del parámetro a y el segundo es una lista
        con el valor del error cuadrático medio en cada iteración. 
    """
    # TODO: Completa la función. Sigue los siguientes pasos:
    # Durante cada iteración:
    #    - Calcula el gradiente de la función de costo con respecto al parámetro a.
    #    - Actualiza el valor de a.
    #    - Obten la predicción del modelo.
    #    - Calcula el error cuadrático medio.
    #    - Guarda el valor de la función de error en la lista a regresar.
    mse_history = []  # Lista para almacenar el error cuadrático medio en cada iteración.

    for _ in range(iterations):
        # Calcula el gradiente de la función de costo con respecto al parámetro a.
        gradient = grad_mse(x, y_data, a)

        # Actualiza el valor de a utilizando el descenso del gradiente.
        a -= learning_rate * gradient

        # Obtiene la predicción del modelo.
        y_pred = model(x, a)

        # Calcula el error cuadrático medio.
        mse_ = mse(y_data, y_pred)

        # Guarda el valor del error en la lista a regresar.
        mse_history.append(mse_)

    return a, mse_history


# Ejemplo de uso de la función
initial_a = 0.5  # Valor inicial aleatorio para a
optimal_a, mse_history = optimize(x, y_data, initial_a)
print(f"initial_a: {initial_a}")
print(f"\nValor optimizado de a: {optimal_a}")
```
